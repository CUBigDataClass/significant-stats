```python
import pandas as pd

df_1 = pd.read_csv("emission data.csv")
df_1.dropna(inplace=True)

df_2 = pd.DataFrame()
df_2['country']='0'
df_2['year']='0'
df_2['emission']='0'

columns = list(df_1)
columns.remove('Country')

for i in range(len(df_1)):
    for j in columns:
        df_2 = df_2.append({'country': df_1['Country'].iloc[i], 'year': str(j), 'emission': df_1[j][i]}, ignore_index=True)

print(df_2)
df_2.to_csv("cleaned_data_emission_data.csv")
```
